```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json

data_folder = os.path.abspath(os.path.join("..", "data"))
print("Files in data folder:", os.listdir(data_folder))

def get_file_path(filename):
    path = os.path.join(data_folder, filename)
    if not os.path.exists(path):
        raise FileNotFoundError(f"File not found: {path}")
    return path

movies_main_path = get_file_path("movies_main.csv")
ratings_path = get_file_path("ratings.json")
movie_extended_path = get_file_path("movie_extended.csv")
```

```python
if check_file(ratings_path):
    print("\nExploring ratings.json...\n")
    with open(ratings_path, "r") as f:
        ratings_data = json.load(f)
    print("Type of ratings_data:", type(ratings_data))
    if isinstance(ratings_data, list):
        print("Sample entry:")
        display(ratings_data[0])
    elif isinstance(ratings_data, dict):
        for k, v in ratings_data.items():
            print(f"{k}: type {type(v)}")
    try:
        ratings_df = pd.json_normalize(ratings_data)
        display(ratings_df.head())
        print("\nMissing Values per Column in Ratings:")
        print(ratings_df.isnull().sum())
        print("\nDuplicate Rows in Ratings:", ratings_df.duplicated().sum())
    except Exception as e:
        print("Could not convert ratings JSON to DataFrame:", e)
```

```python
if check_file(movie_extended_path):
    print("\nExploring movie_extended.csv...\n")
    movie_extended = pd.read_csv(movie_extended_path)
    display(movie_extended.head())
    print("Shape (rows, columns):", movie_extended.shape)
    print("\nMissing Values per Column:")
    print(movie_extended.isnull().sum())
    print("\nDuplicate Rows:", movie_extended.duplicated().sum())
    for col in ['genres', 'production_company', 'country']:
        if col in movie_extended.columns:
            print(f"\nSample {col} values:")
            display(movie_extended[col].head())
```
